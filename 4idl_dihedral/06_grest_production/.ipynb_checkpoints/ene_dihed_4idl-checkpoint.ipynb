```julia
using Revise
using MDToolbox, PyPlot, Statistics, LinearAlgebra
PyPlot.plt.style.use("seaborn-colorblind")
ENV["COLUMNS"] = 110; #display width for MDToolbox
```

```julia
s = readlines("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_all.ene")
```

```julia
N = 3000 #100000 dihed3000 トラジェクトリファイルのフレーム数
STEP = zeros(Int,N)
ENERGY_1 = zeros(Float32,N)
#ENERGY_2 = zeros(Float32,N)
#ENERGY_3 = zeros(Float32,N)
#ENERGY_4 = zeros(Float32,N)
#ENERGY_5 = zeros(Float32,N)
#ENERGY_6 = zeros(Float32,N)
#ENERGY_7 = zeros(Float32,N)
#ENERGY_8 = zeros(Float32,N)
```

```julia
k = 0
for i = 1:length(s) #iは全体のエネルギーの探索
    if endswith(s[i],"#") #ここからはじまり
        for j = 1:100 #N #jは10000000ステップの中での探索(エネルギーは100ステップおきに出力)　このエネルギーを1000でストライドしたトラジェクトリに対応するように抽出していく
            k = k + 1
            STEP[k] = parse(Int,s[i+j*1000][8:14])               #最も左のカラム
            ENERGY_1[k] = parse(Float32,s[i+j*1000][23:34])      #左から2番目のカラム
            #ENERGY_2[j] = parse(Float32,s[i+j][39:50])      #左から3番目のカラム
            #ENERGY_3[j] = parse(Float32,s[i+j][55:66])      #左から4番目のカラム
            #ENERGY_4[j] = parse(Float32,s[i+j][71:82])      #左から5番目のカラム
            #ENERGY_5[j] = parse(Float32,s[i+j][87:98])      #左から6番目のカラム
            #ENERGY_6[j] = parse(Float32,s[i+j][103:114])    #左から7番目のカラム
            #ENERGY_7[j] = parse(Float32,s[i+j][119:130])    #左から8番目のカラム
            #ENERGY_8[j] = parse(Float32,s[i+j][135:146])    #左から9番目のカラム
        end
    end
end
```

```julia
fig, ax = subplots(figsize=(10, 5));
ax.scatter(STEP,ENERGY_1, s=2.0);
xlim([0, 100000])
#ylim([0, 130])
xlabel("step",fontsize=20);
ylabel("Potential Energy",fontsize=20);
xticks(fontsize=10);
yticks(fontsize=10);

#savefig("b.png", dpi=350, bbox_inches="tight");
```

```julia
pdb_xtal = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/ground_truth.pdb");
pdb_init = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/prot.pdb");
ta1 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/all_1000.nc");
ta = [pdb_init; ta1]
pdb_xtal = pdb_xtal["(atomname CA) and (resid 100:114)"]
ta = ta["(atomname CA) and (resid 100:114)"]
ta_fitted = superimpose(pdb_xtal, ta);
rmsd = compute_rmsd(pdb_xtal, ta_fitted)
```

```julia
rmsd_N = zeros(Float32,N)
for i = 1:2950 #N
    rmsd_N[i] = rmsd[i]
end
```

```julia
fig, ax = subplots(figsize=(10, 5));
ax.scatter(rmsd_N,ENERGY_1, s=2.0);
#ax.scatter(rmsd_N,ENERGY_2, s=2.0);
#ax.scatter(rmsd_N,ENERGY_3, s=2.0);
#ax.scatter(rmsd_N,ENERGY_4, s=2.0);
#ax.scatter(rmsd_N,ENERGY_5, s=2.0);
#ax.scatter(rmsd_N,ENERGY_6, s=2.0);
#ax.scatter(rmsd_N,ENERGY_7, s=2.0);
#ax.scatter(rmsd_N,ENERGY_8, s=2.0);
xlim([1.5, 8])
#ylim([-82000, -79100])
xlabel("RMSD",fontsize=20);
ylabel("Potential Energy",fontsize=20);
xticks(fontsize=10);
yticks(fontsize=10);

#savefig("b.png", dpi=350, bbox_inches="tight");
```
