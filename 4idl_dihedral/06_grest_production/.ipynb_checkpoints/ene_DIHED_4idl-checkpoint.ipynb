```julia
N = 9990 #10000に近づける #13000 #トラジェクトリファイルのフレーム数
STEP = zeros(Int,N)
ENERGY_1 = zeros(Float32,N)
ENERGY_2 = zeros(Float32,N)
ENERGY_3 = zeros(Float32,N)
ENERGY_4 = zeros(Float32,N)
ENERGY_5 = zeros(Float32,N)
ENERGY_6 = zeros(Float32,N)
ENERGY_7 = zeros(Float32,N)
ENERGY_8 = zeros(Float32,N)
```

```julia
k = 0
for i = 1:length(s1) #iは全体のエネルギーの探索
    if startswith(s1[i]," --------------- --------------- --------------- --------------- ---------------") #ここからはじまり
        for j = 1:333 #N #jは10000000ステップの中での探索(エネルギーは100ステップおきに出力)　このエネルギーを100でストライドしたトラジェクトリに対応させる
            k = k + 1
            STEP[k] = parse(Int,s1[i+j*200*3-1][8:16])     
            ENERGY_1[k] = parse(Float32,s1[i+j*200*3-1][53:64])
            ENERGY_2[k] = parse(Float32,s2[i+j*200*3-1][53:64]) 
            ENERGY_3[k] = parse(Float32,s3[i+j*200*3-1][53:64]) 
            ENERGY_4[k] = parse(Float32,s4[i+j*200*3-1][53:64]) 
            ENERGY_5[k] = parse(Float32,s5[i+j*200*3-1][53:64]) 
            ENERGY_6[k] = parse(Float32,s6[i+j*200*3-1][53:64]) 
            ENERGY_7[k] = parse(Float32,s7[i+j*200*3-1][53:64]) 
            ENERGY_8[k] = parse(Float32,s8[i+j*200*3-1][53:64]) 
        end
    end
end
```

```julia
#RMSDの計算
pdb_xtal = readpdb("/data/higashi/new_4idl/like_ALL_4IDL/ground_truth.pdb");
pdb_xtal = pdb_xtal["(atomname CA) and (resid 100:114)"];
pdb_init = mdload("/data/higashi/new_4idl/like_ALL_4IDL/prot.pdb");

ta1 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_param1_all_s100.nc");
ta2 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_param2_all_s100.nc");
ta3 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_param3_all_s100.nc");
ta4 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_param4_all_s100.nc");
ta5 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_param5_all_s100.nc");
ta6 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_param6_all_s100.nc");
ta7 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_param7_all_s100.nc");
ta8 = mdload("/data/higashi/grest_4idl_swissmodel_DIHEDRAL/06_grest_production/run_param8_all_s100.nc");

TA1 = [pdb_init; ta1]
TA1 = TA1["(atomname CA) and (resid 100:114)"]
ta_fitted_1 = superimpose(pdb_xtal, TA1);
rmsd1 = compute_rmsd(pdb_xtal, ta_fitted_1)

TA2 = [pdb_init; ta2]
TA2 = TA2["(atomname CA) and (resid 100:114)"]
ta_fitted_2 = superimpose(pdb_xtal, TA2);
rmsd2 = compute_rmsd(pdb_xtal, ta_fitted_2)

TA3 = [pdb_init; ta3]
TA3 = TA3["(atomname CA) and (resid 100:114)"]
ta_fitted_3 = superimpose(pdb_xtal, TA3);
rmsd3 = compute_rmsd(pdb_xtal, ta_fitted_3)

TA4 = [pdb_init; ta4]
TA4 = TA4["(atomname CA) and (resid 100:114)"]
ta_fitted_4 = superimpose(pdb_xtal, TA4);
rmsd4 = compute_rmsd(pdb_xtal, ta_fitted_4)

TA5 = [pdb_init; ta5]
TA5 = TA5["(atomname CA) and (resid 100:114)"]
ta_fitted_5 = superimpose(pdb_xtal, TA5);
rmsd5 = compute_rmsd(pdb_xtal, ta_fitted_5)

TA6 = [pdb_init; ta6]
TA6 = TA6["(atomname CA) and (resid 100:114)"]
ta_fitted_6 = superimpose(pdb_xtal, TA6);
rmsd6 = compute_rmsd(pdb_xtal, ta_fitted_6)

TA7 = [pdb_init; ta7]
TA7 = TA7["(atomname CA) and (resid 100:114)"]
ta_fitted_7 = superimpose(pdb_xtal, TA7);
rmsd7 = compute_rmsd(pdb_xtal, ta_fitted_7)

TA8 = [pdb_init; ta8]
TA8 = TA8["(atomname CA) and (resid 100:114)"]
ta_fitted_8 = superimpose(pdb_xtal, TA8);
rmsd8 = compute_rmsd(pdb_xtal, ta_fitted_8)
```

```julia
rmsd_N_1 = zeros(Float32,N) #zeros(Float32,N)
rmsd_N_2 = zeros(Float32,N) #zeros(Float32,N)
rmsd_N_3 = zeros(Float32,N) #zeros(Float32,N)
rmsd_N_4 = zeros(Float32,N) #zeros(Float32,N)
rmsd_N_5 = zeros(Float32,N) #zeros(Float32,N)
rmsd_N_6 = zeros(Float32,N) #zeros(Float32,N)
rmsd_N_7 = zeros(Float32,N) #zeros(Float32,N)
rmsd_N_8 = zeros(Float32,N) #zeros(Float32,N)

for i = 1:N #サイズ合わせる
    rmsd_N_1[i] = rmsd1[3*i]
    rmsd_N_2[i] = rmsd2[3*i]
    rmsd_N_3[i] = rmsd3[3*i]
    rmsd_N_4[i] = rmsd4[3*i]
    rmsd_N_5[i] = rmsd5[3*i]
    rmsd_N_6[i] = rmsd6[3*i]
    rmsd_N_7[i] = rmsd7[3*i]
    rmsd_N_8[i] = rmsd8[3*i]
end
```
